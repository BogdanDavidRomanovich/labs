# Определение полиномов

```python
# Полином Чебышева первого рода (PCPR)

def PCPR(x, p):
    if p == 0:
        return 1
    elif p == 1:
        return x
    else:
        return 2 * x * PCPR(x, p - 1) - PCPR(x, p - 2)
```

```python
# Смещенный полином Чебышева (SPCPR)

def SPCPR(x, p):
    if p == 0:
        return 1
    elif p == 1:
        return -1 + 2 * x
    else:
        return 2 * (-1 + 2 * x) * SPCPR(x, p - 1) - SPCPR(x, p - 2)
```

```python
# Полином Чебышева второго рода (PCVR)

def PCVR(x, p):
      if p == 0:
        return 1
      elif p == 1:
        return 2 * x
      else:
        return 2 * x * PCVR(x, p - 1) - PCVR(x, p - 2)
```

```python
# Смещенный полином Чебышева второго рода (SPCVR)

def SPCVR(x, p):
      if p == 0:
        return 1
      elif p == 1:
        return 4 * x - 2
      else:
        return 2 * (2 * x - 1) * SPCVR(x, p - 1) - SPCVR(x, p - 2)
```

```python
# Полином Лежандра (LNDR)

def LNDR(x, p):
      if p == 0:
        return 1
      elif p == 1:
        return x
      else:
        return ((2 * p - 1) * x * LNDR(x, p - 1) - (p - 1) * LNDR(x, p - 2)) / p
```

```python
# Полином Лагерра (LGR)

def LGR(x, p):
      if p == 0:
        return 1
      elif p == 1:
        return 1 - x
      else:
        return (2 * p - 1 - x) * LGR(x, p - 1) - (p - 1) ** 2 *LGR(x, p - 2)
```

```python
# Полином Эрмита (HERM)

def HERM(x, p):
      if p == 0:
        return 0
      elif p == 1:
        return x
      else:
        x * HERM(x, p - 1) - (p - 1) * HERM(x, p - 2)
```

# Ввод данных




```python
# Объявление пустых списков для хранения входных данных

X1 = []
X2 = []
X3 = []
Y = []
```

```python
# Считывание данных из файлов Х1, Х2, Х3 и Y

X1txt = open("X1.txt","r")
for l in X1txt:
    X1.append([float(x) for x in l.split()])
    
X2txt = open("X2.txt","r")
for l in X2txt:
      X2.append([float(x) for x in l.split()])

X3txt = open("X3.txt","r")
for l in X3txt:
      X3.append([float(x) for x in l.split()])

Ytxt = open("Y.txt", "r")
for l in Ytxt:
      Y.append([float(x) for x in l.split()])
```

```python
n1 = len(X1[0])  # n1 - размерность X1
n2 = len(X2[0])  # n2 - размерность Х2
n3 = len(X3[0])  # n3 - размерность Х3
n = len(Y)       # n  - объем выборки 
m = len(Y[0])    # m  - размерность Y
```

```python
# сохранение считанных данных в виде матриц

X1 = np.array(X1)
X2 = np.array(X2)
X3 = np.array(X3)
Y = np.array(Y)
```

```python
# Создание таблицы входных данных

InpDF = pd.DataFrame({
}, index = np.arange(n) + 1)

for j in range(n1):
      InpDF['X1' + str(j + 1)] = X1[:, j]
        
for j in range(n2):
      InpDF['X2' + str(j + 1)] = X2[:, j]
        
for j in range(n3):
      InpDF['X3' + str(j + 1)] = X3[:, j]
        
for j in range(m):
      InpDF['Y' + str(j + 1)] = Y[:, j]
```

```python
# Выбор полинома 

fi = SPCVR   # Для того, чтобы изменить вид полинома, замените слово справа от символа "=" на условное обозначение желаемого   

# Возможные альтернативы:
# Смещенный полином Чебышева (SPCPR)
# Смещенный полином Чебышева второго рода (SPCVR)
# Полином Лежандра (LNDR)
# Полином Лагерра (LGR)
# Полином Эрмита (HERM)
```

```python
# Степени полиномов

P1 = 20
P2 = 20
P3 = 20
```

# Нормирование данных

```python
maxX1 = np.array([max(X1[:, j]) for j in range(n1)])
minX1 = np.array([min(X1[:, j]) for j in range(n1)])
  
maxX2 = np.array([max(X2[:, j]) for j in range(n2)])
minX2 = np.array([min(X2[:, j]) for j in range(n2)])

maxX3 = np.array([max(X3[:, j]) for j in range(n3)])
minX3 = np.array([min(X3[:, j]) for j in range(n3)])

maxY = np.array([max(Y[:, j]) for j in range(m)])
minY = np.array([min(Y[:, j]) for j in range(m)])

for j in range(n1):
    X1[:, j] = (X1[:, j] - minX1[j]) / (maxX1[j] - minX1[j])
    
for j in range(n2):
    X2[:, j] = (X2[:, j] - minX2[j]) / (maxX2[j] - minX2[j])

for j in range(n3):
      X3[:, j] = (X3[:, j] - minX3[j]) / (maxX3[j] - minX3[j])

for j in range(m):
      Y[:, j] = (Y[:, j] - minY[j]) / (maxY[j] - minY[j])
```

# Определение коефициентов при полиномах - матриц  $||\lambda||$


$b_{iq_0}$ принимаются равными нормированным значениям $Y_i [q_0]$

```python
b = Y.copy()
```

## Способ №1

```python
# Создаем пустую матрицу L коэффициентов при неизвестных лямбдах

L = np.empty((n, n1 * (P1 + 1) + n2 * (P2 + 1) + n3 * (P3 + 1)))
L.shape     # размерность матрицы L
```

```python
# Заполняем L значениями 

for j1 in range(n1):
      for p1 in range(P1 + 1):
        L[:, p1 + j1 * (P1 + 1)] = fi(X1[:, j1], p1)
    
for j2 in range(n2):
      for p2 in range(P2 + 1):
        L[:, p2 + j2 * (P2 + 1) + n1 * (P1 + 1)] = fi(X2[:, j2], p2)
    
for j3 in range(n3):
      for p3 in range(P3 + 1):
        L[:, p3 + j3 * (P3 + 1) + n1 * (P1 + 1) + n2 * (P2 + 1)] = fi(X3[:, j3], p3) 
```

```python
# Решаем систему с помощью МНК для каждого bi (b = [bi] i = 1,...)

Lambda = np.linalg.lstsq(L, b)[0]
Lambda.shape  
```

# Определение структуры векторов $x_1, x_2, x_3 $ - матриц $||a||$

```python
Psi1 = np.empty((m, n, n1))
Psi2 = np.empty((m, n, n2))
Psi3 = np.empty((m, n, n3))
```

```python
for i in range(m):
  
          Psi1[i, :, :] = np.array([[(L * Lambda[:, i])[x, j1 * (P1 + 1) : (j1 + 1) * (P1 + 1)].sum() for j1 in range(n1)] for x in range(n)])

          Psi2[i, :, :] = np.array([[(L * Lambda[:, i])[x, n1 * (P1 + 1) + j2 * (P2 + 1) : n1 * (P1 + 1) + (j2 + 1) * (P2 + 1)].sum() for j2 in range(n2)] for x in range(n)])
        
          Psi3[i, :, :] = np.array([[(L * Lambda[:, i])[x, n1 * (P1 + 1) + n2 * (P2 + 1) + j3 * (P3 + 1): n1 * (P1 + 1) + n2 * (P2 + 1) + (j3 + 1) * (P3 + 1)].sum() for j3 in range(n3)] for x in range(n)])
```

```python
A1 = np.empty((m, n1))
A2 = np.empty((m, n2))
A3 = np.empty((m, n3))
```

```python
for i in range(m):
        A1[i] = np.linalg.lstsq(Psi1[i], b[:, i])[0]
        A2[i] = np.linalg.lstsq(Psi2[i], b[:, i])[0]
        A3[i] = np.linalg.lstsq(Psi3[i], b[:, i])[0]
```

# Определение структуры векторов функций $Ф_i$ -матриц $||c||$

```python
Fi1 = np.empty((m, n))
Fi2 = np.empty((m, n))
Fi3 = np.empty((m, n))
```

```python
for i in range(m):
    Fi1[i] = Psi1[i] @ A1[i]
    Fi2[i] = Psi2[i] @ A2[i]
    Fi3[i] = Psi3[i] @ A3[i]
```

```python
F = np.empty((m, 45, 3))
```

```python
F[0].shape
```

```python
for i in range(m):
      F[i] = np.array([Fi1[i], Fi2[i], Fi3[i]]).transpose()
```

```python
C = np.empty((m, 3))
```

```python
for i in range(m):
      C[i] = np.linalg.lstsq(F[i], b[:, i])[0]
```

```python
Ya = np.array([F[i] @ C[i] for i in range(m)])
for i in range(m):
      Ya[i] = Ya[i] * (maxY[i] - minY[i]) + minY[i] 
```
